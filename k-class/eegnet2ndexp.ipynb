```python
# Lib
import sys
sys.path.append("/home/rffl/files/eegconv/")

from torch.utils.data import DataLoader
import eegnet # type: ignore
import kclassgen
import kclassdataset

# Classes
classvec = classvec = [i for i in range(10)]
```

## Dataset and dataloader from train data

```python
train_path, test_path = kclassgen.generatedataset(classvec, 3, 0.75, True)

trainDataset = kclassdataset.KClassDataset(len(classvec), train_path)
trainLoader = DataLoader(trainDataset, 12, True, num_workers=3)

testDataset = kclassdataset.KClassDataset(len(classvec), test_path)
testLoader = DataLoader(testDataset, 12, True, num_workers=3)
```

### Training and Testing

```python
eegnet.regular_train(17, 100, 100, len(classvec), trainLoader, testLoader, 100)
```

## n-Subject training
This next section encompasses the training with multiple subjects.

### Dataset and dataloader instantiation from testing data

```python
import kclass_lsubgen # type: ignore

classvec = [3, 6, 9]
subvec = [i+1 for i in range(5)]

train_path, test_path = kclass_lsubgen.generatedataset(classvec, subvec, 0.75, False)

trainDataset = kclassdataset.KClassDataset(len(classvec), train_path)
trainLoader = DataLoader(trainDataset, 12, True, num_workers=3)

testDataset = kclassdataset.KClassDataset(len(classvec), test_path)
testLoader = DataLoader(testDataset, 12, True, num_workers=3)
```

### Training and testing

```python
# Cross = False (p_dropout = 0.5)
eegnet.regular_train(17, 100, 100, len(classvec), trainLoader, testLoader, 100, False, classvec)
```

```python
# Cross = True (p_dropout = 0.25)
eegnet.regular_train(17, 100, 100, len(classvec), trainLoader, testLoader, 100, True, classvec)
```
