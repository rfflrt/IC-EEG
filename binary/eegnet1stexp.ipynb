# First Explroation of EEGNet on THINGS-EEG
This notebook is the first exploration of my implementatio of EEGNet on THINGS-EEG dataset. <br>
This will test the convergence of the loss on a dataset of binary data from two classes. <br>
Also, the EEGs will be from one subject only.

```python
# Lib
import sys
sys.path.append("/home/rffl/files/eegconv/binary/")

from torch.utils.data import DataLoader
import eegnet
import bindatagen
import bindataset 
```

## Dataset and dataloader instantiation

```python
trainDataset = bindataset.BinEEGDataset(bindatagen.generatetrain(1,2))
trainLoader = DataLoader(trainDataset, 5, True, num_workers=2)

testDataset = bindataset.BinEEGDataset(bindatagen.generatetest(1,2))
testLoader = DataLoader(testDataset, 5, True, num_workers=2)
```

## Network instantiation

```python
eegnet.regular_train(17, 100, 100, 2, trainLoader, testLoader, 25)
```
